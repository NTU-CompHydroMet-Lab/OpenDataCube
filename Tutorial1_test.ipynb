```python
datasetsNew = dc.find_datasets(
    product="era5_1",
    time=('2022-01-01', '2022-01-02')
)

ds = dc.load(
    datasets=datasetsNew,
    measurements=["radar"],
    output_crs="EPSG:4326",
    resolution=(-1,1),
)
ds
```
